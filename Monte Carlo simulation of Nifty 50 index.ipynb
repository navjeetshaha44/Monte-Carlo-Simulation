```python
#Monte Carlo simulation of Nifty 50 index:

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


no_of_days = 252

simulation_runs = 100000
spot_price = 15000
strike = 15000
barrier = 1500
vol = 0.15

np.random.seed(25)
rets = np.random.randn(simulation_runs,no_of_days)*vol/np.sqrt(252)
rets.shape

lines = np.cumprod(1+rets,1)*spot_price
for i in lines[:100,:]:
    plt.plot(i)
    
plt.title("Possible Paths", fontsize=18)

plt.ylabel('Price', fontsize=15)
plt.xlabel('Year', fontsize=15)

plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
